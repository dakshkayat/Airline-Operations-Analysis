# PART A

## 1 What is the total cost by aircraft type for the year?

```python
op=operations
```

```python
df_grouped = op.groupby('Aircraft Type').sum(numeric_only=True)

# Calculate the total hours for each aircraft type
df_total_hours = df_grouped.sum(axis=1).reset_index()  # Resetting index for a clean DataFrame

# Rename columns to match the desired output
df_total_hours.columns = ['Aircraft_Type', 'Total_hours_flown']
```

```python
df_total_hours
```

## 2 Which aircraft type has the lowest cost per seat per km flown?

```python
df_joined = ac_characteristics.merge(df_total_hours, left_on='Aircraft_Type', right_on='Aircraft_Type')
```

```python
df_joined
```

```python
df_joined['Total_cost'] = df_joined['Costs_per_flight_hour'] * df_joined['Total_hours_flown']
df_joined['Total_distance'] = df_joined['Total_hours_flown'] * df_joined['Average_Speed (km/h)']
```

```python
df_joined["cost per seat per km"]=df_joined["Total_cost"] / (df_joined["Number_of_Seats"] * df_joined["Total_distance"])
```

# PART B

## Which aircraft types are best suited for their operation? 

```python
result = pd.merge(ac_characteristics.assign(key=1), city_pairs.assign(key=1), on='key').drop('key', axis=1)
result 
```

```python
import math

# Calculate required trips
result['Required Trips'] = result['Pass. Demand(per day)'] / result['Number_of_Seats']
result['Required Trips'] = result['Required Trips'].apply(math.ceil)

result
```

```python
# Filter by range
filter_data=result[(result['Range (Km)'] >= result['Distance (km)'])]
filter_data
```

```python
filter_data['Total Cost']=(filter_data['Distance (km)']/filter_data['Average_Speed (km/h)']*filter_data['Costs_per_flight_hour'])*filter_data['Required Trips']
```

```python
filter_data
```

```python
# Group by city pairs and select the row with the minimum total cost for each pair
best_aircraft_for_city_pairs = filter_data.loc[filter_data.groupby(['Origin City', 'Desitnation City'])['Total Cost'].idxmin()]

best_aircraft_for_city_pairs[['Origin City','Desitnation City', 'Aircraft_Type']]
```
